```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob

def plot_simulation_data(file_path=None, columns_to_plot=None, title=None, log_scale=False):
    """
    Plot simulation data from a CSV file.
    
    Args:
        file_path (str): Path to the CSV file. If None, uses the most recent file.
        columns_to_plot (list): List of column names to plot. If None, plots all percentage columns.
        title (str): Title for the plot. If None, uses a default title.
        log_scale (bool): Whether to use logarithmic scale for y-axis.
    
    Returns:
        None
    """
    # If no file specified, find the most recent one
    if file_path is None:
        csv_files = glob.glob('data/simulation_data_*.csv')
        if not csv_files:
            print("No simulation data CSV files found in the data directory")
            return
        file_path = max(csv_files)
    
    print(f"Loading data from: {file_path}")
    
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # Print basic info about the data
    print(f"Data contains {len(df)} time points")
    print(df.head())
    
    # If no columns specified, use all percentage columns
    if columns_to_plot is None:
        columns_to_plot = [col for col in df.columns if 'percent' in col]
    
    # Define colors for different cell types
    colors = {
        'cancer_percent': 'r',
        'wbc_percent': 'g',
        'dead_percent': 'k',
        'normal_percent': 'b',
        'regen_percent': 'm',
        'ctt_percent': 'y'
    }
    
    # Create a figure with appropriate size
    plt.figure(figsize=(12, 8))
    
    # Plot each specified column against timestamp
    for column in columns_to_plot:
        if column in df.columns:
            color = colors.get(column, 'gray')  # Default to gray if column not in colors dict
            label = column.replace('_percent', '').capitalize() + ' %'
            plt.plot(df['timestamp'], df[column], f'{color}-', label=label)
    
    # Add labels and title
    plt.xlabel('Timestamp', fontsize=12)
    
    if log_scale:
        plt.ylabel('Percentage (log scale)', fontsize=12)
        plt.yscale('log')
    else:
        plt.ylabel('Percentage', fontsize=12)
        plt.ylim(0, 100)
    
    # Set default title if none provided
    if title is None:
        title = 'Cell Population Dynamics Over Time'
        if log_scale:
            title += ' (Log Scale)'
    
    plt.title(title, fontsize=14)
    
    # Add legend and grid
    plt.legend(fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Show the plot
    plt.tight_layout()
    plt.show()

# Example usage:
# Plot all percentages with linear scale

```

```python
# BEGIN: Modify plot_multiple_simulation_data to calculate slopes
def plot_multiple_simulation_data(file_paths, attribute='cancer_percent', labels=None):
    """
    Plot a specified attribute from multiple CSV files over time and calculate slopes between y=20 and y=70.
    
    Args:
        file_paths (list): List of paths to the CSV files.
        attribute (str): The attribute to plot. Defaults to 'cancer_percent'.
        labels (list): Custom labels for each file. If None, uses file names.
    
    Returns:
        None
    """
    all_data = []
    
    for file_path in file_paths:
        print(f"Loading data from: {file_path}")
        df = pd.read_csv(file_path)
        
        if attribute in df.columns:
            all_data.append(df[['timestamp', attribute]])
        else:
            print(f"Warning: {attribute} not found in {file_path}")
    
    # Concatenate all dataframes
    combined_df = pd.concat(all_data, ignore_index=True)
    
    # Create a figure
    plt.figure(figsize=(12, 8))
    
    # Define colors for different files
    colors = plt.cm.viridis(np.linspace(0, 1, len(file_paths)))
    
    # Plot each dataframe and calculate slopes
    for i, file_path in enumerate(file_paths):
        df = pd.read_csv(file_path)
        if labels and len(labels) > i:
            label = labels[i]
        else:
            label = file_path.split('/')[-1]
        
        if attribute in df.columns:
            plt.plot(df['timestamp'], df[attribute], label=label, color=colors[i])
            
            # Calculate slope between y=20 and y=70
            y_values = df[attribute]
            x_values = df['timestamp']
            indices = np.where((y_values >= 20) & (y_values <= 80))[0]
            if len(indices) > 1:
                slope = (y_values[indices[-1]] - y_values[indices[0]]) / (x_values[indices[-1]] - x_values[indices[0]])
                print(f"Slope for {label}: {slope}")
    
    # Add labels and title
    plt.xlabel('Timestamp', fontsize=20)
    plt.ylabel(attribute.replace('_', ' ').capitalize(), fontsize=20)
    plt.title(f'{attribute.replace("_", " ").capitalize()} Over Time', fontsize=20)
    
    # Add legend and grid
    plt.legend(fontsize=20)
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Show the plot
    plt.tight_layout()
    plt.show()
# END:
```

```python
plot_multiple_simulation_data(
    file_paths=[
        'data/simulation_data_125.csv',
                'data/simulation_data_162.csv',

        'data/simulation_data_224.csv',
        'data/simulation_data_457.csv',
        'data/simulation_data_550.csv',

        'data/simulation_data_1138.csv'
    ],
    labels=[
        'Cancer transformation prob: 0.2', 
                'Cancer transformation prob: 0.15', 

    'Cancer transformation prob: 0.1', 
    'Cancer transformation prob: 0.05',
        'Cancer transformation prob: 0.04'

    ,'Cancer transformation prob: 0.02']
)
```

```python
# Slope for Cancer transformation prob: 0.2: 1.0454385964912278
# Slope for Cancer transformation prob: 0.1: 0.5615330188679246
# Slope for Cancer transformation prob: 0.05: 0.30146464646464644
# Slope for Cancer transformation prob: 0.04: 0.24410204489795917
# Slope for Cancer transformation prob: 0.02: 0.12045271629778671
# Prepare data for plotting
slopes = [
    1.0454385964912278,  # prob: 0.2
    0.7973332666666666,  # prob: 0.15
    0.5615330188679246,  # prob: 0.1
    0.30146464646464644, # prob: 0.05
    0.24410204489795917, # prob: 0.04
    0.12045271629778671  # prob: 0.02
]
probabilities = [0.2,0.15, 0.1, 0.05, 0.04, 0.02]
k, b = np.polyfit(probabilities, slopes, 1)

# Plotting the relationship
plt.figure(figsize=(10, 6))
plt.plot(probabilities, slopes, marker='o', linestyle='-', color='b')
plt.xlabel('Cancer Transformation Probability', fontsize=20)
plt.ylabel('Slope (s)', fontsize=20)
plt.title('Relationship between Probability and Slope', fontsize=20)
plt.grid(True)
plt.text(0.05, 0.8 * max(slopes), f'k = {k:.4f} * p + {b:.4f}', fontsize=20, color='red')

plt.show()

# Inferring the relationship s = kp
# Assuming a linear relationship, we can fit a line to the data
print(f"Fitted line: s = {k:.4f} * p + {b:.4f}")
```

```python
import pandas as pd
import numpy as np

def find_cancer_wbc_at_zero_slope(file_path):
    """
    Reads a CSV file and finds the values of cancer_percent and wbc_percent
    when the slope of cancer_percent approaches zero.

    Args:
        file_path (str): The path to the CSV file.

    Returns:
        list: A list of tuples containing (cancer_percent, wbc_percent) 
              when the slope is close to zero.
    """
    df = pd.read_csv(file_path)
    
    if 'cancer_percent' not in df.columns or 'wbc_percent' not in df.columns:
        print("Required columns are not present in the CSV file.")
        return []

    results = []
    y_values = df['cancer_percent'].values
    x_values = np.arange(len(y_values))

    # Calculate the slope using numpy's gradient
    slopes = np.gradient(y_values)

    # Find indices where the slope is close to zero
    zero_slope_indices = np.where(np.isclose(slopes, 0, atol=3e-3))[0]
    # BEGIN: Plot slopes against indices

    indices = np.arange(len(slopes))
    plt.figure(figsize=(10, 6))
    plt.plot(indices, slopes, marker='o', linestyle='-', color='b')
    plt.xlabel('Index', fontsize=12)
    plt.ylabel('Slope', fontsize=12)
    plt.title('Relationship between Index and Slope', fontsize=14)
    plt.grid(True)
    plt.show()
    # END:
    for index in zero_slope_indices:
        # print(slopes[index])
        print(index)
        results.append((df['cancer_percent'].iloc[index], df['wbc_percent'].iloc[index]))

    return results
```
